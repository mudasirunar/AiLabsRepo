```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = pd.read_csv('homeprices.csv')

X = data[['area']] 
y = data['price']  
model = LinearRegression()

model.fit(X, y)
predicted_prices = model.predict(X)

print("Original and Predicted Prices:")
output = pd.DataFrame({'Area': X['area'], 'Original Price': y, 'Predicted Price': predicted_prices})
print(output)

plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, predicted_prices, color='red', label='Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Linear Regression: Area vs Price')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.DataFrame({
    'area': [2600, 3000, 3200, 3600, 4000],
    'price': [550000, 565000, 610000, 680000, 725000]
})

X = data[['area']]
y = data['price'] 

model = LinearRegression()
model.fit(X, y)

predicted_areas = [[5000], [8000], [9000]] 
predicted_prices = model.predict(predicted_areas)

for area, price in zip(predicted_areas, predicted_prices):
    print(f"Predicted price for a home with area {area[0]} sq ft: ${price:.2f}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = pd.read_csv('carprices.csv')

X = data[['model']]  
y = data['price']   

model = LinearRegression()
model.fit(X, y)

predicted_prices = model.predict(X)

print("Original and Predicted Prices:")
output = pd.DataFrame({'Model': X['model'], 'Original Price': y, 'Predicted Price': predicted_prices})
print(output)

plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, predicted_prices, color='red', label='Regression Line')
plt.xlabel('Car Model (Year)')
plt.ylabel('Price')
plt.title('Linear Regression: Car Model vs Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
data = pd.DataFrame({
    'model': [2019, 2020, 2021, 2022, 2023, 2024],
    'price': [1500000, 2000000, 3000000, 3500000, 4000000, 4500000]
})

X = data[['model']]  
y = data['price']    

model = LinearRegression()
model.fit(X, y)

future_models = [[2025], [2026], [2027]]  
predicted_prices = model.predict(future_models)

for year, price in zip(future_models, predicted_prices):
    print(f"Predicted price for a car with model year {year[0]}: ${price:.2f}")
```
